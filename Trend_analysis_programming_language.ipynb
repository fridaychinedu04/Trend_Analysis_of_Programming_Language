```python
# IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# LOADING THE DATASET
prog_lang = pd.read_csv(r"c:\Users\DELL\Desktop\stack_overflow_data.csv")
prog_lang
```

```python
# Question 1: COMPUTE PROPORTION OF QUESTIONS PER LANGUAGE PER YEAR
prog_lang["proportion"] = prog_lang["num_questions"] / prog_lang["year_total"]
print(prog_lang.head(), "\n")
```

```python
# Question 2: TOP FIVE MOST POPULAR PROGRAMMING LANGUAGE EACH YEAR
top5 = (prog_lang.sort_values(["year","proportion"], ascending=[True, False]).groupby("year").head(5).reset_index(drop=True))
print(top5)
```

```python
# Question 3: PLOT TREND OF FIVE SELECTED LANGUAGES
selected_languages = ["python","java","javascript","c#","c++"]
plot_trend = prog_lang[prog_lang["tag"].isin(selected_languages)]
plt.figure(figsize=(12,6))
for lang in selected_languages:
    lang_data = plot_trend[plot_trend["tag"] == lang]
    plt.plot(lang_data["year"],lang_data["proportion"],label=lang)
plt.title("Trend of Programming Language Popularity (2008 - 2020)")
plt.xlabel("Year")
plt.ylabel("Proportion of Question")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Question 4: SHOW TREND TABLE FOR SELECTED LANGUAGES
trend_table = (plot_trend[["year","tag","proportion"]].sort_values(["tag","year"]).reset_index(drop=True))
print(trend_table)
```
